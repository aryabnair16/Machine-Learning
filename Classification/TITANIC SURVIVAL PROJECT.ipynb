The sinking of the Titanic is one of the most infamous shipwrecks in history.

On April 15, 1912, during her maiden voyage, the widely considered “unsinkable” RMS Titanic sank after colliding with an iceberg. Unfortunately, there weren’t enough lifeboats for everyone onboard, resulting in the death of 1502 out of 2224 passengers and crew.

While there was some element of luck involved in surviving, it seems some groups of people were more likely to survive than others.

In this challenge, we ask you to build a predictive model that answers the question: “what sorts of people were more likely to survive?” using passenger data (ie name, age, gender, socio-economic class, etc).

Data Dictionary:
   * survived - (0 - No, 1 - Yes)
   * pclass - Ticket class
   * Sex - Sex
   * Age - Age in Years
   * Sibsp - No: of siblings/spouses aboard the Titanic
   * Parch - No: of parents/children aboard the Titanic
   * Ticket - Ticket Number
   * Fare - Passenger Fare
   * Cabin - Cabin Number
   * Embarked - Port of embarkation


### Import necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

## Clean the dataset

### Remove unwanted features (PassengerId,Name, ticket, Cabin)

```python
df.drop(columns = ['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace = True)
df.head()
```

### Fill null values in Age with mean



```python
df['Age'].fillna(df['Age'].mean(), inplace = True)
```

```python
df.isna().sum()
```

### Fill null values in Embarked with 'Unknown'

```python
df['Embarked'].fillna('Unknown', inplace = True)
```

```python
df.isna().sum()
```

### Convert the categorical datas into numerical ('Sex', 'Embarked')

```python
df = pd.get_dummies(df, drop_first= True)
df.head()
```

## Split the set into feature and target varibles(X,y)

```python
X = df.drop(columns = ['Survived'])
y = df['Survived']
```

### Check the shape of X and y

```python
X.shape
```

```python
y.shape
```

### Standardise the data with standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xcolumns = X.columns
```

```python
X = scaler.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns = xcolumns)
```

```python
X.head()
```

## <span style = "color: green"> Visualization </span>

### Plot a pair plot

```python
sns.pairplot(df)
plt.plot()
```

### Plot a heat map to view the correlation between features

```python
plt.figure(figsize = (10,8))
sns.heatmap(df.corr(), annot = True)
plt.plot()
```

### Split the data into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state= 100)
```

#### Check the shape of X_train and X_test

```python
X_train.shape
```

```python
X_test.shape
```

### Create a Logistic Regression model and Train it

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
# Train the model
model.fit(X_train, y_train)
```

### Check the score of our model

```python
model.score(X_train,y_train)
```

### Make predictions using X_test

```python
y_pred = model.predict(X_test)
y_pred
```

### Check the accuracy score

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,y_pred)
```

### Check the confusion metrics

```python
metrics.confusion_matrix(y_test,y_pred)
```

### Plot confusion matrix

```python
sns.heatmap(metrics.confusion_matrix(y_test,y_pred), annot = True, fmt = 'd')
plt.show()
```

### Print the classification report

```python
print(metrics.classification_report(y_test,y_pred))
```
